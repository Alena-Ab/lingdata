## ЧАСТЬ 3: создание и преобразование столбцов

#### Создайте колонку 'Top_University' со значением True для университетов, входящих в топ-50. Обратите внимание на значения типа "231–240".

```python
df['University Global Ranking'].unique()  #проверка данных
```

```python
def ranking_to_num(ranking):
    if '–' in str(ranking):
        return None
    else:
        return float(ranking) #убираем лишнее


df['University Global Ranking'] = df['University Global Ranking'].apply(ranking_to_num)
df['University Global Ranking'].unique()
```

```python
df['Top_University'] = df['University Global Ranking'] <= 50
df.head()
```

```python
df.shape
```

#### Создайте колонку Academic_Performance, в которую внесите значение:
#### - "High", если GPA выше 3.5 баллов,
#### - "Medium", если GPA от 3 до 3.49 баллов,
#### - "Low", если GPA ниже 3 баллов.

```python
def gpa_value(value):
    if value > 3.5:
        return 'High'
    elif value < 3:
        return 'Low'
    else:
        return 'Medium'


df['Academic_Performance'] = df['GPA (or Equivalent)'].apply(gpa_value)
df.head()
```

#### Посмотрите на уникальные значения в колонке Degree и приведите их к единообразию (Bachelor of Science -> Bachelor).

```python
df.shape
```

```python
df['Degree'].value_counts()  #проверка данных (+количества каждого degree)
```

```python
def degrees(degrees):
  if ('Master' in str(degrees) or 'MD' in str(degrees) or 'MBA' in str(degrees)): #"MBA and BA" я отнесла к Master как к общему, тк нельзя ведь получить магистра без бакалавра (поэтому и код начала с Master)
        return 'Master'
  elif 'Bachelor' in str(degrees):
          return 'Bachelor'
  elif ('Ph' in str(degrees) or 'Doctor' in str(degrees)):
          return 'Doctor'
  elif 'Drop' in str(degrees):
          return 'Dropped Out'
  elif 'Engineering' in str(degrees):
          return 'Not specified'
  elif 'Specialist' in str(degrees):
          return 'Specialist'
  return degrees #приводим к приблизительно общему виду (specialist degree я поменяла на specialist ради красоты, можно было этого и не делать)

df_degrees = pd.read_csv("successful_educations.csv") #на всякий случай новый датасет

df_degrees['Degree'] = df['Degree'].apply(degrees)
df_degrees['Degree'].value_counts()
```
