## ЧАСТЬ 3: создание и преобразование столбцов

#### Создайте колонку 'Top_University' со значением True для университетов, входящих в топ-50. Обратите внимание на значения типа "231–240".

#### Создайте колонку Academic_Performance, в которую внесите значение:
#### - "High", если GPA выше 3.5 баллов,
#### - "Medium", если GPA от 3 до 3.49 баллов,
#### - "Low", если GPA ниже 3 баллов.

```python
def gpa_value(value):
    if value >= 3.5:
        return 'High'
    elif value < 3:
        return 'Low'
    else:
        return 'Medium' #сортируем значения


df['Academic_Performance'] = df['GPA (or Equivalent)'].apply(gpa_value) #новая колонка
df.head() #проверяем
```

#### Посмотрите на уникальные значения в колонке Degree и приведите их к единообразию (Bachelor of Science -> Bachelor).

```python
df['Degree'].value_counts()  #проверка данных (+количества каждого degree)
```

```python
def degrees(degrees):
  if ('Master' in str(degrees) or 'MD' in str(degrees) or 'MBA' in str(degrees)): #"MBA and BA" я отнесла к Master как к общему, тк нельзя ведь получить магистра без бакалавра (поэтому и код начала с Master)
        return 'Master'
  elif 'Bachelor' in str(degrees):
          return 'Bachelor'
  elif ('Ph' in str(degrees) or 'Doctor' in str(degrees)):
          return 'Doctor'
  elif 'Drop' in str(degrees):
          return 'Dropped Out'
  elif 'Engineering' in str(degrees):
          return 'Not specified'
  elif 'Specialist' in str(degrees):
          return 'Specialist'
  return degrees #приводим к приблизительно общему виду (specialist degree я поменяла на specialist ради красоты)

df_degrees = pd.read_csv("successful_educations.csv") #на всякий случай новый датасет

df_degrees['Degree'] = df['Degree'].apply(degrees)
df_degrees['Degree'].value_counts()
```

```python
df.head(2) #смотрим названия нужных столбцов и их формат
```

```python
df_alumni = pd.read_csv("successful_educations.csv") #новый датасет

df_alumni['Graduation Year'].unique()  #проверяем
```

```python
def years(year):
    if 'USA(residence)/ Hungary (origin)' in str(year):
        return None
    if pd.isna(year):
        return None
    return str(year).strip() #убираем лишнее

df_alumni['Graduation Year'] = df_alumni['Graduation Year'].apply(years)

df_alumni = df_alumni.dropna(subset=['Graduation Year']) #удаляем NA в столбце с годом выпуска

df_alumni['Graduation Year'].unique()  #проверяем
```

```python
df_alumni['Graduation Year'] = pd.to_datetime(df_alumni['Graduation Year'], format='%Y')  #меняем тип на datetime64

df_alumni = df_alumni.set_index('Graduation Year')  #делаем года выпуска индексами
df_alumni.head(2)  #проверяем
```

```python
df_alumni.resample('10YE').size()  #группировка по десятилетиям и количество выпускников
```

```python
df_alumni.resample('10YE')['Field'].value_counts().groupby(level=0).head(1)
#наиболее популярная область обучения для каждого десятилетия
#(level=0).head(1) подглядела у дипсика, потому что не знала, как отобразить только самую поп область
#(без этого показывает количество людей на каждую область в каждом десятилетии)
```

```python
df_alumni['GPA (or Equivalent)'] = df_alumni['GPA (or Equivalent)'].apply(gpa_to_num) #берем код оттуда, где чистили столбец с гпа
df_alumni['GPA (or Equivalent)'] = df_alumni['GPA (or Equivalent)'].fillna(df['GPA (or Equivalent)'].mean()) #заполняем столбец с гпа средним значением в новом дф

df_alumni.resample('10YE').agg({'GPA (or Equivalent)': 'mean'}) #считаем среднее
```
